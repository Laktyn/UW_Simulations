```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

# Zadanie 1

```python
# time step function

def change(row):
    next_row = row.copy()
    row_left = np.roll(row.copy(), -1)
    row_right = np.roll(row.copy(), 1)
    next_row = row_left + row_right
    next_row[next_row == 3] = 1
    next_row[next_row == 2] = 0
    return next_row

# initialize rows

random_row = np.random.choice(np.array([0,1]), 501)
not_random_row = np.concatenate([np.zeros(230), np.ones(41), np.zeros(230)])

# create full lattice

lattice = [not_random_row,]
row = not_random_row

for time in range(1000):
    row = change(row)
    lattice.append(row)

plt.imshow(lattice)

width = 501
height = 1000
img = Image.new("RGB", (width,height), (255, 255, 255))
draw = ImageDraw.Draw(img)
for y in range(height):
    for x in range(width):
        if lattice[y][x]: draw.point((x, y), (0, 0, 0))
img.save(fp = "Lab11.1 - struktura nielosowa.png", format = "PNG")
```

# Zadanie 1 - v2

```python
def automat(row1, row2, row3, number):
    bin_num = bin(number)[2:]
    vector = []
    for i in range(8 - len(bin_num)):
        vector.append(0)
    for i in bin_num:
        vector.append(int(i))
    together = np.array(4*row1 + 2*row2 + 1*row3, dtype = int)
    return np.array(vector)[7-together]

row = np.concatenate([np.zeros(230), np.ones(41), np.zeros(230)])

lattice = [row,]
for time in range(1000):
    if time % 10 == 0:
        print(str(time/10) + "%")
    row = automat(np.roll(row, -1), row, np.roll(row, 1), number = 110)
    print(len(row))
    lattice.append(row)

width = 501
height = 1000
img = Image.new("RGB", (width,height), (255, 255, 255))
draw = ImageDraw.Draw(img)
for y in range(height):
    if y % 10 == 0:
        print(str(y/10) + "%")
    for x in range(width):
        if lattice[y][x]: draw.point((x, y), (0, 0, 0))
img.save(fp = "Lab11.1 - regua 35.png", format = "PNG")
```
